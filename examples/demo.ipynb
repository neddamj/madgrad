```python
import sys
sys.path.append('..')

import numpy as np
from sklearn.datasets import make_moons, make_blobs, make_circles
import matplotlib.pyplot as plt

import mdgrad
import mdgrad.nn as nn
import mdgrad.optim as optim
```

```python
X, targets = make_moons(n_samples=100, noise=0.1)
targets = np.expand_dims(targets, axis=-1)
X, targets = mdgrad.tensor(X), mdgrad.tensor(targets)
# visualize in 2D
plt.figure(figsize=(5,5))
plt.scatter(X[:,0], X[:,1], c=targets.numpy(), s=20, cmap='jet')
```

```python
X.shape, targets.shape
```

```python
model = nn.Sequential(
    nn.Linear(2, 20),
    nn.ReLU(),
    nn.Linear(20, 50), 
    nn.ReLU(),
    nn.Linear(50, 15),
    nn.ReLU(),
    nn.Linear(15, 1),
    nn.Sigmoid()
)
```

```python
def accuracy(x, y):
    x = mdgrad.tensor([[1 if pred > 0.5 else 0 for pred in x]])
    y = y.reshape(x.shape)
    return mdgrad.sum(x.data == y.data) / 100
```

```python
lr = 0.5
loss_fn = nn.MSELoss()
optimizer = optim.SGD(model.parameters(), lr=lr, momentum=0.9)

losses, accs = [], []
model.train()
for i in range(50):
    preds = model(X)
    loss = loss_fn(preds, targets)
    acc = accuracy(preds, targets)
    if (i+1) % 10 == 0 or i == 0:
        print(f'Epoch: {i+1}\tLoss: {loss.item():.4f}\tAccuracy: {100 * acc.item():.2f}%')
    
    # Optimize the model
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

    # track model performance
    losses.append(loss.item())
    accs.append(acc.item())

model.eval()
```

```python
plt.plot(losses, label='MSE Loss')
plt.plot(accs, label='Accuracy')
plt.xlabel('Epochs')
plt.legend()
plt.grid()
```
