```python
import numpy as np
from sklearn.datasets import make_moons, make_blobs, make_circles
import matplotlib.pyplot as plt

import madnet as mn
import madnet.nn as nn
import madnet.nn.optim as optim
```

```python
X, targets = make_moons(n_samples=100, noise=0.1)
targets = np.expand_dims(targets, axis=-1)
X, targets = mn.Tensor(X), mn.Tensor(targets)
# visualize in 2D
plt.figure(figsize=(5,5))
plt.scatter(X[:,0], X[:,1], c=targets.data, s=20, cmap='jet')
```

```python
X.shape, targets.shape
```

```python
model = nn.Sequential([
    nn.Linear(2, 20),
    nn.ReLU(),
    nn.Linear(20, 50), 
    nn.ReLU(),
    nn.Linear(50, 15),
    nn.ReLU(),
    nn.Linear(15, 1),
    nn.Sigmoid()
])
```

```python
lr = 0.5
loss_fn = nn.MSELoss()
optimizer = optim.SGD(model.parameters(), lr=lr)
```

```python
def accuracy(x, y):
    x = np.array([[1 if pred > 0.5 else 0 for pred in x]])
    y = y.reshape(x.shape)
    return np.sum(x == y) / 100
```

```python
losses, accs = [], []

for i in range(100):
    preds = model(X)
    loss = loss_fn(preds, targets)
    acc = accuracy(preds, targets)
    if (i+1) % 10 == 0 or i == 0:
        print(f'Epoch: {i+1}\tLoss: {loss.data[0][0]:.4f}\tAccuracy: {100 * acc:.2f}%')
    # Optimize the model
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

    # track model performance
    losses.append(loss.data[0][0])
    accs.append(acc)
```

```python
plt.plot(losses, label='MSE Loss')
plt.plot(accs, label='Accuracy')
plt.xlabel('Epochs')
plt.legend()
plt.grid()
```
